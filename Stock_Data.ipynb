```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
tickers_list={"AAPL","NVDA","MSFT","AMZN","GOOGL","META","TSLA","TSM","AVGO","WMT","LLY","JPM","V","ORCL","UNH","XOM","NVO","MA","COST"}
stock_data=pd.DataFrame()
for tickers in tickers_list:
 stock_data[tickers]=yf.download(tickers,period="10y")["Adj Close"]
```
